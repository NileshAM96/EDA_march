**Data Transformation**
- Generally data has 3 types of distributions
    -Normal distribution
    - positive skewed
    - Negative skewed
- All the maths concepts are developed by assuming that the data follows normal distribution
- Data skew happens because of many reasons, in that one reason is outliers
- Once we perform the outlier analysis then we can apply transformation methods if the data does not follow normality
- We have some transformation methods mentiomed below
        - Log transformation
        - Exponential transformation
        - Reciprocol Transformation
        - Square root transformation
        - Power transformation



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
exp_data = np.random.exponential(size = 1000)
exp_data
```

```python
#step -3 = plot the ddaata
plt.hist(exp_data, bins = 20, label = "Exponential")
```

```python
norm_data = np.random.normal(size = 1000)
norm_data
plt.hist(norm_data, bins = 20, label = "Exponential")
```

```python
# our goal is to convert exp data to normal data
```

# step4 - Log transformation
- Log transformation means performing logarithm operations on original data
- It is an approach to convert into Normal distribution
- Log means natural logarithm means base = e


```python
log_data = np.log(exp_data)
log_data
plt.subplot(1,2,1).hist(exp_data, bins =30, label = 'Exponential')
plt.legend()
plt.subplot(1,2,2).hist(log_data, bins = 20, label = "Logarithmic")
plt.legend()
```

**Reciprocol Transformations**

```python
reciprocol_data = np.reciprocal(exp_data)
reciprocol_data
plt.subplot(1,2,1).hist(exp_data, bins =30, label = 'Exponential')
plt.legend()
plt.subplot(1,2,2).hist(reciprocol_data, bins = 20, label = "Reciprocol")
plt.legend()

# reciprocol will fail when values are low
# for example x=0 then 1/0 fail
```

```python

sqrt_data = np.sqrt(exp_data)
sqrt_data
plt.subplot(1,2,1).hist(exp_data, bins =30, label = 'Exponential')
plt.legend()
plt.subplot(1,2,2).hist(sqrt_data, bins = 20, label = "SquareRoot")
plt.legend()
```
