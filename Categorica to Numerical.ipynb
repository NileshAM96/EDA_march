#np.where() method
- np.where is applicable only for binary
- np.where will take 3 arguments -
    - condition
    - True value
    - False value

$scenario$ 
    - for example case status value is equal to Certified replace with zero
    - Otherwise replace with 1

```python
visa_df=pd.read_csv(r"Visadataset - Visadataset.csv")
visa_df.head(2)
```

```python
con = visa_df['case_status']== 'Certified'
visa_df['case_status'] = np.where(con,0,1)
visa_df
```

#One hot encoder

- One hot means if one is ON another one is OFF
- ON represents with 1 and OFF represents with 0
- For example case status has two unique labels
    - Denied
    - Certified
- If we apply one hot encoder on case_status column it will create two extra columns
    - case_status_Denied
    - case_status_Certified


**Advantages**
- In machine learning we have one important property is **Independent of variables**
- Machine learning models expects variables are **independent each other**
- One hot encoder creating new columns, the new columns are **independent of each other**
- Independent variable means both varianles have **90 degrees phase shift**
- 90 degrees phase shift also called as **perpendicular to each other**
- Perpendicular means **orthogonal to each other**

**Disadvantages**
- One hot encoder creates new columns equal to number of unique labels
- If number of unique labels increase, new columns also increase
- The more columns lead data is more
- The data is more the complexity will increase
- Complexity increases more computation power required
- Also time also increase
- The entire process is named as **curse of dimensionality**
